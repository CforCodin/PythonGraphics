OR Gate

```python
import pandas as pd
import numpy as nm
import matplotlib.pyplot as plt
x = nm.array([0,0,1,1])
y = nm.array([0,1,0,1])
S = nm.array([0,1,1,1])
X = nm.array([1,2,3,4])
d = pd.DataFrame({'A':x, 'B':y, 'C':S},index=['*','*','*','*'])
plt.plot(X,x,c='red',marker='o')
plt.plot(X,y,c='yellow',marker='o')
plt.plot(X,S,c='blue',marker='o')
plt.title("OR Gate")
print(d)
```

```python
import pandas as pd
import numpy as nm
import matplotlib.pyplot as plt
x = nm.array([0,0,1,1])
y = nm.array([0,1,0,1])
S = nm.array([0,0,0,1])
X = nm.array([1,2,3,4])
d = pd.DataFrame({'A':x, 'B':y, 'C':S},index=['*','*','*','*'])
plt.plot(X,x,c='red',marker='o')
plt.plot(X,y,c='yellow',marker='o')
plt.plot(X,S,c='blue',marker='o')
plt.title("AND Gate")
print(d)
```
